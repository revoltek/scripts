```python
from lib_analysis_lolss import *
%matplotlib inline

cm_inferno = plt.colormaps['inferno']

cat_lolss = Table.read('/homes/fdg/storage/surveytgts/catalogues/LoLSS_DR2-cat.fits')
cat_lolss.rename_column('Total_flux', 'flux_lolss')
cat_lolss.rename_column('Peak_flux', 'peak_lolss')
cat_lolss.rename_column('E_Total_flux', 'flux_lolss_e')

cat_nvss = Table.read('/homes/fdg/storage/catalogues/NVSS.fits')
cat_nvss.rename_column('Total_flux', 'flux_nvss')
cat_nvss['flux_nvss'] = cat_nvss['flux_nvss']/1e-3 # Jy -> mJy

cat_lotss = Table.read('/homes/fdg/storage/catalogues/LoTSS_DR3_v1.0.srl.fits')
cat_lotss.rename_column('Total_flux', 'flux_lotss')
cat_lotss.rename_column('Peak_flux', 'peak_lotss')
cat_lotss.rename_column('E_Total_flux', 'flux_lotss_e')

# select compact for lotss
snr = cat_lotss['flux_lotss']/cat_lotss['flux_lotss_e']
R = 0.42+(1.08/( 1 + (snr/96.57)**(2.49) ))
cat_lotss['is_pnt'] = (R > np.log(cat_lotss['flux_lotss']/cat_lotss['peak_lotss'])) 
print('Is pnt (LoTSS) %f%%' % (100*sum(cat_lotss['is_pnt'])/len(cat_lotss)))

cat_lolss_lotss = cross_match_multicomp(cat_lolss, cat_lotss, dist=15, flux_col='flux_lotss', columns_to_keep=['is_pnt'])
print('Match LoTSS: %i' % sum(np.isfinite(cat_lolss_lotss['flux_lotss_sum'])))
print('un-Match LoTSS: %i' % sum(np.isnan(cat_lolss_lotss['flux_lotss_sum'])))
cat_lolss_lotss['spidxLoLSSLoTSS'] = np.log10(cat_lolss_lotss['flux_lotss_sum']/cat_lolss_lotss['flux_lolss'])/np.log10(144/54.)
cat_lolss_lotss.write('/homes/fdg/storage/catalogues/cat_lolss-lotss.fits', overwrite=True)

# select compact for lolss
snr = cat_lolss['flux_lolss']/cat_lolss['flux_lolss_e']
R = 0.14+(1.02/( 1 + (snr/161.)**(1.25) ))
cat_lolss['is_pnt'] = (R > np.log(cat_lolss['flux_lolss']/cat_lolss['peak_lolss'])) 
print('Is pnt (LoLSS) %f%%' % (100*sum(cat_lolss['is_pnt'])/len(cat_lolss)))

cat_lolss_nvss = cross_match_multicomp(cat_nvss, cat_lolss, dist=45, flux_col='flux_lolss', columns_to_keep=['is_pnt'])
print('Match NVSS: %i' % sum(np.isfinite(cat_lolss_nvss['flux_lolss_sum'])))
print('un-Match NVSS: %i' % sum(np.isnan(cat_lolss_nvss['flux_lolss_sum'])))
cat_lolss_nvss['spidxLoLSSNVSS'] = np.log10(cat_lolss_nvss['flux_nvss']/cat_lolss_nvss['flux_lolss_sum'])/np.log10(1400/54.)
cat_lolss_nvss.write('/homes/fdg/storage/catalogues/cat_lolss-nvss.fits', overwrite=True)

#cat_lolss_nvss_lotss = cross_match_multicomp(cat_lolss_nvss, cat_lotss, dist=45, flux_col='flux_lotss', columns_to_keep=['spidxLOW'])
#print('Match LoTSS+NVSS: %i' % sum(np.isfinite(cat_lolss_nvss_lotss['flux_lotss_sum'])))
#print('un-Match LoTSS+NVSS: %i' % sum(np.isnan(cat_lolss_nvss_lotss['flux_lotss_sum'])))
#cat_lolss_nvss_lotss['spidxLoTSSNVSS'] = np.log10(cat_lolss_nvss_lotss['flux_nvss']/cat_lolss_nvss_lotss['flux_lotss_sum'])/np.log10(1400/144.)
#cat_lolss_nvss_lotss.write('/homes/fdg/storage/catalogues/cat_lolss-nvss-lotss.fits', overwrite=True)
```
